```python
from tkinter import *
import sqlite3

#Create database and table
connection = sqlite3.connect('ExamScore.db')
c = connection.cursor()
c.execute("""CREATE TABLE IF NOT EXISTS StudentsScores(ID INTEGER PRIMARY KEY, FullName text, Score text)""")
connection.commit()


#Function that takes text from boxes and adds them to database
def addData():
    fullname = fullNameBox.get()
    score = scoreBox.get();
    c.execute("""INSERT INTO StudentsScores(FullName, Score) 
                VALUES(?, ?)""", (fullNameBox.get(), scoreBox.get()))
    connection.commit()
    



#Functiom that display data from database
def displayScores():
    c.execute("""SELECT * FROM StudentsScores ORDER BY Score DESC""")
    fullnamesAndScores = c.fetchall()

    for i in fullnamesAndScores:
        displayBox.insert(END, fullnamesAndScores[0])
    
    
    

    
#Function that clears everything from boxes
def clearBoxes():
    fullNameBox.delete(0, END)
    scoreBox.delete(0, END)
    displayBox.delete(0, END)
    
    



window = Tk()
window.geometry('650x650')
window.title("ExamScores")

#Create labels and text place for full name
fullNameLabel = Label(text="Enter student's full name")
fullNameLabel.place(x = 10, y = 10, height = 25, width = 300)
fullNameBox = Entry(text = "")
fullNameBox.place(x = 300, y = 10, width = 300, height = 25)

#Create labels and text place for score
scoreLabel = Label(text="Enter student's Exam score")
scoreLabel.place(x = 10, y = 50, height = 25, width = 300)
scoreBox = Entry(text = "")
scoreBox.place(x = 300, y = 50, width = 300, height = 25)

#Create Add button
addButton = Button(text="Add", command = addData)
addButton.place(x = 350, y = 80, width = 50, height = 25)

#Create Clear button
clearButton = Button(text="Clear", command = clearBoxes)
clearButton.place(x = 430, y = 80, width = 50, height = 25)


#Create Display label, button and box
displayLabel = Label(text = "Display all students' record: ")
displayLabel.place(x = 150, y = 210, height = 25, width = 300)
displayBox = Listbox()
displayBox.place(x = 200, y = 250, height = 300, width = 400)
displayButton = Button(text="Display", command = displayScores)
displayButton.place(x = 90, y =350, height = 25, width = 70)




window.mainloop()
```
